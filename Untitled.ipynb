## Creating Count Vectorizer
Possible to use TFIDF Vectorizer instead and use it
[countvectorizer and tfidf vectorizer](https://machinelearningmastery.com/prepare-text-data-machine-learning-scikit-learn/)

```python
from sklearn.feature_extraction.text import CountVectorizer

#Creating the bag of words
bow_title = CountVectorizer().fit(news['title'])
```

```python
title_vect = bow_title.transform(news['title'])
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer().fit(title_vect)
news_tfidf = tfidf_transformer.transform(title_vect)
print(news_tfidf.shape)
```

```python
X = news_tfidf
y = news['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test= train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.svm import SVC
model = SVC()
detector = model.fit(X_train, Y_train)
```

```python
predictions = detector.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import classification_report
print (classification_report(Y_test, predictions))
```

```python
detector.predict(tfidf_transformer.transform(bow_title.transform(["You Can Smell Hillary’s Fear"])))
```

# Using Text Body instead of Title

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectorizer.fit(news['text'])
text_vector = vectorizer.transform(news['text'])
```

```python
X = text_vector
y = news['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test= train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.svm import SVC
model = SVC()
detector = model.fit(X_train, Y_train)
```

```python
predictions = detector.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import classification_report
print (classification_report(Y_test, predictions))
```

# Conclusion

As we can observe, predicting weather the news is fake or real is much accurate when using text body as compared with when using news title itself.
